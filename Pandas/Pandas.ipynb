```python
import pandas as pd
# Pandas Series Object -> Series Object is one-dimensional labeled object
s1 = pd.Series([1,2,3,4])
s1
```

```python
type(s1)
```

```python
# changing Index
s1 = pd.Series([1,2,3,4],index = ['a','b','c','d'])
s1
```

```python
# Series Object from Dictionary
s1 = pd.Series({'a':10 , 'b':20 , 'c':30})
s1
```

```python
# changing index position
s1 = pd.Series({'a':10 , 'b':20 ,'c':30} , index = ['a','c','b','d'])
s1
```

```python
# Extracting individual Element
s1 = pd.Series([1,2,3,4,5,6,7,8])
s1[3]
```

```python
# Extracting a series of Elements
s1[5:]
```

```python
# Extracting element from a Back
s1[-3:]
```

```python
# Basci Maths operation on Series
s1+4
```

```python
s2 = pd.Series([10,11,12,13,14,15,16,17])
s1+s2
```
